### Listly를 활용한 크롤링

Listly  프로그램을 활용하여 네이버 부동산 각 역별로 전체 아파트 매물을 크롤링 하여 기초 데이터 확보

```python
import pandas as pd
seoksu = pd.read_excel("/home/has/Downloads/석수역.xlsx")
seoksu
```

```python
seoksu=seoksu.iloc[:,:14]
seoksu=seoksu.drop(['LABEL-2','LABEL-5','LABEL-3','LABEL-7','LABEL-8','LABEL-9','LABEL-10','LABEL-11','LABEL-12','LABEL-13'],axis=1)
seoksu=seoksu.rename(columns = {'LABEL-1':'아파트명','LABEL-4':'매매가','LABEL-6':'면적','LABEL-14':'링크'})
seoksu=seoksu.dropna(axis=0)
seoksu
```

```python
seoksu['면적']=seoksu.면적.str.split('㎡').str[0]
seoksu['공급면적']=seoksu.면적.str.split('/').str[0]
seoksu['전용면적']=seoksu.면적.str.split('/').str[1]
seoksu=seoksu.drop(['면적'],axis=1)
seoksu['매매가']=seoksu.매매가.str.split('~').str[0]
seoksu
```

```python
seoksu=seoksu.astype({ '공급면적' : 'int' })
seoksu=seoksu.astype({ '전용면적' : 'int' })
seoksu['전용률']=seoksu['전용면적']/seoksu['공급면적']*100
seoksu['전용률']= round(seoksu['전용률'],0)
seoksu
```

```python
p1=seoksu.매매가.str.split('억 ').str[1]
p2=seoksu.매매가.str.split('억').str[0]
p1=p1.fillna(0)
p2=p2.astype('int')
p2=p2*100000000
p1=p1.str.replace(pat = r'[^\w]', repl=r'', regex=True)
p1=p1.fillna(0)
p1=p1.astype('int')
p1=p1*10000
seoksu['매매가']=p2+p1
```

```python
seoksu['평수']=seoksu['전용면적']/3.3
seoksu['평당가격']=seoksu['매매가']/seoksu['평수']/10000
seoksu['평수']= round(seoksu['평수'],1)
seoksu['평당가격']=round(seoksu['평당가격'],0)
seoksu
```

### 석수역 아파트 가격

```python
seoksu= seoksu.sort_values(by=['평당가격'], ascending=True)
seoksu=seoksu.drop(['링크'],axis=1)
seoksu
```
